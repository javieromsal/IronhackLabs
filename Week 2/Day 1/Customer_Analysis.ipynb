## Filtering data and Correcting typos – Filter the data in state and gender column to standardize the texts in those columns

```python
df['state'].value_counts()
```

```python
def typos(x):
    if x in ['Cali']:
        return 'California'
    elif x in ['AZ']:
        return 'Arizona'
    elif x in ['WA']:
        return 'Washington'
    else:
        return x
    
```

```python
df['state'] = list(map(typos, df['state']))
```
