### Cleaning and wrangling data

- Some car brand names refer to the same brand. Replace all brand names that contain the word "Dutton" for simply "Dutton". If you find similar examples, clean their names too. Use `loc` with boolean indexing.

- Convert CO2 Emissions from Grams/Mile to Grams/Km

- Create a binary column that solely indicates if the transmission of a car is automatic or manual. Use `pandas.Series.str.startswith` and .

- convert MPG columns to km_per_liter

Note:
<br>Converting Grams/Mile to Grams/Km

1 Mile = 1.60934 Km

Converting Gallons to Liters

1 Gallon = 3.78541 Liters



```python
vehicles_df[vehicles_df["Make"].str.contains("Dutton")] #Find the ones that contain "Dutton"
```

```python
vehicles_df.loc[vehicles_df["Make"].str.contains('Dutton'),"Make"]="Dutton"
```

```python
vehicles_df.iloc[vehicles_df["Make"].str.contains('Dutton'), 0] = 'Dutton' #cambia las que contegan 'Dutton a solo Dutton'
vehicles_df[vehicles_df["Make"].str.contains("Dutton")]
```

```python
vehicles_df.loc[vehicles_df["Make"] == "Dutton" ]
```

```python
vehicles_df["CO2 Emission Grams/Mile"] = vehicles_df["CO2 Emission Grams/Mile"]*(1/1.60934)
vehicles_df
```

```python
vehicles_df.rename(columns={"CO2 Emission Grams/Mile":"CO2 Emissions Grams/Kilometer"},inplace=True)
```

**Create a binary column that solely indicates if the transmission of a car is automatic or manual. Use pandas.Series.str.startswith**

```python
vehicles_df.loc[vehicles_df['Transmission'].str.startswith('M'),'Manual/Automatic'] = 1
vehicles_df.loc[vehicles_df['Transmission'].str.startswith('A'),'Manual/Automatic'] = 0
vehicles_df
```

```python
vehicles_df.columns
```

```python
vehicles_df["City MPG"]=vehicles_df["City MPG"]*(1.60934/3.78541)
vehicles_df["Highway MPG"]=vehicles_df["Highway MPG"]*(1.60934/3.78541)
vehicles_df["Combined MPG"]=vehicles_df["Combined MPG"]*(1.60934/3.78541)

vehicles_df = vehicles_df.rename(columns={"City MPG":"City Km/L","Highway MPG":"Highway Km/L","Combined MPG":"Combined Km/L"})

vehicles_df



```
